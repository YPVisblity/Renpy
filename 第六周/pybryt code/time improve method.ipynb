```python
import pybryt

algo_ref = []
recursive_ref = []

def binarySearch_algorithm(arr, l, r, x, track=True):
    while l <= r:
        m = (l + r) // 2
        if track:
            algo_ref.append(pybryt.Value(m))
        if arr[m] == x:
            return m
        if arr[m] < x:
            l = m + 1
        else:
            r = m - 1
    return -1

def binarySearch_recursive(arr, l, r, x, track=True):
    if l <= r:
        mid = (l + r) // 2
        if track:
            recursive_ref.append(pybryt.Value(mid))
        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            return binarySearch_recursive(arr, l, mid - 1, x, track=track)
        else:
            return binarySearch_recursive(arr, mid + 1, r, x, track=track)
    return -1

test_arr = [2, 3, 4, 10, 40]
test_x = 10
binarySearch_algorithm(test_arr, 0, len(test_arr) - 1, test_x, track=True)
binarySearch_recursive(test_arr, 0, len(test_arr) - 1, test_x, track=True)


for exp in [2,3,4]: 
    n = 10 ** exp
    large_arr = range(n)
    target = n - 1  

    binarySearch_algorithm(large_arr, 0, n - 1, target, track=True)
    binarySearch_recursive(large_arr, 0, n - 1, target, track=True)

algo = pybryt.ReferenceImplementation("algo", algo_ref, display_name="迭代版二元搜尋")
recursive = pybryt.ReferenceImplementation("recursive", recursive_ref, display_name="遞迴版二元搜尋")
algo.dump("binarysearch_algo.pkl")
recursive.dump("binarysearch_recursive.pkl")

ref1 = pybryt.ReferenceImplementation.load('binarysearch_algo.pkl')
ref2 = pybryt.ReferenceImplementation.load("binarysearch_recursive.pkl")

def student_binary_search(arr, l, r, x):
    while l <= r:
        m = (l + r) // 2
        if arr[m] == x:
            return m
        if arr[m] < x:
            l = m + 1
        else:
            r = m - 1
    return -1
    
with pybryt.check([ref1, ref2]):
    student_binary_search(test_arr, 0, len(test_arr) - 1, test_x)
    for exp in [2,3,4]:
        n = 10 ** exp
        large_arr = range(n)
        target = n - 1
        student_binary_search(large_arr, 0, n - 1, target)
```
