```python
import pybryt
import pybryt.complexities as cplx
import asyncio
import sys

# 解決 Windows 異步與編碼問題
if sys.platform == 'win32':
    asyncio.set_event_loop_policy(asyncio.WindowsSelectorEventLoopPolicy())

algo_ref = []
recursive_ref = []

def binarySearch_algorithm(arr, l, r, x, track=True):
    global pybryt # 確保背景編譯時抓得到模組
    while (l <= r):
        m = (l + r) // 2
        if track:
            algo_ref.append(pybryt.Value(m))
        if (arr[m] == x):
            return m
        if (arr[m] < x):
            l = m + 1
        else:
            r = m - 1
    return -1

def binarySearch_recursive(arr, l, r, x, track=True):
    global pybryt
    if (l <= r):
        mid = (l + r) // 2
        if track:
            recursive_ref.append(pybryt.Value(mid))
        if (arr[mid] == x):
            return mid
        elif (arr[mid] > x):
            return binarySearch_recursive(arr, l, mid - 1, x, track=track)
        else:
            return binarySearch_recursive(arr, mid + 1, r, x, track=track)
    return -1

# 錄製基礎邏輯足跡 (Value Trace)
test_data = [
    ([2, 3, 4, 10, 40], 10),
    ([1, 5, 8, 12, 15, 20, 30], 5)
]

for arr, x in test_data:
    binarySearch_algorithm(arr, 0, len(arr)-1, x, track=True)
    binarySearch_recursive(arr, 0, len(arr)-1, x, track=True)

# 打包成唯一的 Reference 物件供 CLI 使用
full_reference = pybryt.ReferenceImplementation(
    "binary_search_logic", 
    algo_ref + recursive_ref, 
    display_name="二元搜尋邏輯比對"
)
```
