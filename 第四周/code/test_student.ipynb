```python
import pybryt
import asyncio
import sys

# 這是解決 Windows CLI 崩潰的唯一內嵌補丁
if sys.platform == 'win32':
    asyncio.set_event_loop_policy(asyncio.WindowsSelectorEventLoopPolicy())

def student_binary_search(arr, l, r, x):
    while (l <= r):
        m = (l + r) // 2
        if (arr[m] == x): return m
        if (arr[m] < x): l = m + 1
        else: r = m - 1
    return -1

# 必須有觸發測試
test_arr = [2, 3, 4, 10, 40]
student_binary_search(test_arr, 0, len(test_arr)-1, 10)
dummy_ref = pybryt.ReferenceImplementation("dummy", [])
```
