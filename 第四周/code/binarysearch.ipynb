```python
import pybryt
import pybryt.complexities as cplx

# 1. 載入老師的參考答案
ref1 = pybryt.ReferenceImplementation.load('binarysearch_algo.pkl')
ref2 = pybryt.ReferenceImplementation.load("binarysearch_recursive.pkl")

# 2. 讀取學生代碼 (subm.py)
with open("subm.py", "r", encoding="utf-8") as f:
    student_code = f.read()

# 3. 定義一個「手動錄影機」
# 我們利用 PyBryt 的 BaseTracer 來手動抓取記憶體變化
tracer = pybryt.execution.tracing.BaseTracer()

print("===== 🚀 啟動手動側錄系統 =====")

# 開始側錄
with tracer:
    # 執行學生代碼
    exec(student_code)
    
    # 執行邏輯測試
    test_arr = [2, 3, 4, 10, 40]
    test_x = 10
    # 注意：這裡的 function name 必須跟學生寫的一致
    if 'student_binary_search' in locals():
        student_binary_search(test_arr, 0, len(test_arr)-1, test_x)
        
        # 執行效能測試
        for exp in range(5, 12):
            n = 2 ** exp
            large_arr = list(range(n)) 
            target = large_arr[-1]
            with pybryt.check_time_complexity("logn", n):
                student_binary_search(large_arr, 0, len(large_arr)-1, target)
    else:
        print("❌ 錯誤：找不到 student_binary_search 函式")

# 4. 手動提取足跡 (Footprint)
footprint = tracer.get_footprint()

# 5. 直接拿足跡去對答案 (繞過所有自動批改的 wrapper)
res1 = ref1.run(footprint)
res2 = ref2.run(footprint)

print("\n===== 🏆 最終批改報告 (終極穩定版) =====")
print(f"📌 {ref1.display_name}")
print(f"   -> 是否吻合: {res1.correct}")
print(f"   -> 吻合信心度: {res1.confidence * 100:.1f}%")

print(f"\n📌 {ref2.display_name}")
print(f"   -> 是否吻合: {res2.correct}")
print(f"   -> 吻合信心度: {res2.confidence * 100:.1f}%")
```
